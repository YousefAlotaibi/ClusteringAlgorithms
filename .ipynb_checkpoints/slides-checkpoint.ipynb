## Introduction to clustering 
- Clustering is an unsupervised learning method 
- The process of clustering involves grouping the data points into a number of groups so that the data points in each group are more similar to one another and different from the data points in the other groups.

## K-Means Clustering Process
##### 1- assign K number of clusters (it's best to preform the elbow method to determine the best K number of clusters)
##### 2- assign K number of centroids on the graph randomly 
##### 3- calculate the Euclidean distance and assign the points to the nearest centroid


##### euclidean Formula:
 <img src="Images/euclidean.png" width =10%, align = left>
<br>

##### 4- Find the new original centroid in each group
##### 5- Redo step 3 & 4 until the centroids don't change
